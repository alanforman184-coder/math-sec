```python
def caesar_encrypt(text, shift):
    result = ""
    for char in text:
        if char.isalpha():
            if char.isupper():
                result += chr((ord(char) - ord('A') + shift) % 26 + ord('A'))
            else:
                result += chr((ord(char) - ord('a') + shift) % 26 + ord('a'))
        else:
            result += char
    return result
```

```python
def caesar_decrypt(cipher, shift):
    return caesar_encrypt(cipher, -shift)
```

```python
def atbash(text):
    result = ""
    for char in text:
        if char.isalpha():
            if char.isupper():
                result += chr(25 - (ord(char) - ord('A')) + ord('A'))
            else:
                result += chr(25 - (ord(char) - ord('a')) + ord('a'))
        else:
            result += char
    return result
```

```python
def test_examples():
    print("=" * 50)
    print("TESTING CAESAR CIPHER")
    print("=" * 50)

    # Example 1: Julius Caesar (shift=3)
    text1 = "Veni, vidi, vici"
    enc1 = caesar_encrypt(text1, 3)
    dec1 = caesar_decrypt(enc1, 3)
    print(f"\nExample 1 (shift=3):")
    print(f"Original:  {text1}")
    print(f"Encrypted: {enc1}")
    print(f"Decrypted: {dec1}")
    print(f"Success: {text1.lower() == dec1.lower()}")

    # Example 2: Augustus (shift=1)
    text2 = "Festina lente"
    enc2 = caesar_encrypt(text2, 1)
    dec2 = caesar_decrypt(enc2, 1)
    print(f"\nExample 2 (shift=1):")
    print(f"Original:  {text2}")
    print(f"Encrypted: {enc2}")
    print(f"Decrypted: {dec2}")
    print(f"Success: {text2.lower() == dec2.lower()}")

    print("\n" + "=" * 50)
    print("TESTING ATBASH CIPHER")
    print("=" * 50)

    # Example 3: Atbash
    text3 = "ABC XYZ"
    enc3 = atbash(text3)
    dec3 = atbash(enc3)
    print(f"\nExample 3:")
    print(f"Original:       {text3}")
    print(f"Atbash:         {enc3}")
    print(f"Double Atbash:  {dec3}")
    print(f"Success: {text3 == dec3}")
```

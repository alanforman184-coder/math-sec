```python
# Russian alphabet mapping (standard order)
LETTER_TO_NUM = {
    "а": 1, "б": 2, "в": 3, "г": 4, "д": 5, "е": 6, "ё": 7, "ж": 8, "з": 9,
    "и": 10, "й": 11, "к": 12, "л": 13, "м": 14, "н": 15, "о": 16, "п": 17,
    "р": 18, "с": 19, "т": 20, "у": 21, "ф": 22, "х": 23, "ц": 24, "ч": 25,
    "ш": 26, "щ": 27, "ъ": 28, "ы": 29, "ь": 30, "э": 31, "ю": 32, "я": 33
}
```

```python
def text_to_numbers(text):
    """Convert text to list of numbers using the alphabet mapping"""
    text = text.lower()
    numbers = []
    for char in text:
        if char in LETTER_TO_NUM:
            numbers.append(LETTER_TO_NUM[char])
        else:
            # Skip characters not in alphabet (spaces, punctuation)
            continue
    return numbers
```

```python
def numbers_to_text(numbers):
    """Convert list of numbers back to text"""
    text = ""
    for num in numbers:
        if num in NUM_TO_LETTER:
            text += NUM_TO_LETTER[num]
    return text
```

```python
def gamma_encrypt(text, gamma):
    """
    Encrypt text using gamma cipher
    text: plaintext (Russian)
    gamma: key phrase (Russian)
    """
    # Convert text and gamma to numbers
    text_numbers = text_to_numbers(text)
    gamma_numbers = text_to_numbers(gamma)

    print(f"Text numbers: {text_numbers}")
    print(f"Gamma numbers: {gamma_numbers}")

    # Encrypt: (text_num + gamma_num) mod 33
    # Note: mod 33 with numbers 1-33 requires special handling
    result_numbers = []
    gamma_len = len(gamma_numbers)

    for i, text_num in enumerate(text_numbers):
        gamma_num = gamma_numbers[i % gamma_len]  # Repeat gamma if needed
        sum_num = text_num + gamma_num

        # Mod 33 for numbers 1-33
        if sum_num > 33:
            sum_num = sum_num % 33
            if sum_num == 0:
                sum_num = 33

        result_numbers.append(sum_num)

    print(f"Result numbers: {result_numbers}")

    # Convert back to text
    encrypted_text = numbers_to_text(result_numbers)
    return encrypted_text
```

```python
def gamma_decrypt(cipher, gamma):
    """
    Decrypt text using gamma cipher
    cipher: ciphertext (Russian)
    gamma: key phrase (Russian)
    """
    # Convert cipher and gamma to numbers
    cipher_numbers = text_to_numbers(cipher)
    gamma_numbers = text_to_numbers(gamma)

    print(f"Cipher numbers: {cipher_numbers}")
    print(f"Gamma numbers: {gamma_numbers}")

    # Decrypt: (cipher_num - gamma_num) mod 33
    result_numbers = []
    gamma_len = len(gamma_numbers)

    for i, cipher_num in enumerate(cipher_numbers):
        gamma_num = gamma_numbers[i % gamma_len]  # Repeat gamma if needed
        diff_num = cipher_num - gamma_num

        # Mod 33 for numbers 1-33
        if diff_num < 1:
            diff_num = diff_num + 33

        result_numbers.append(diff_num)

    print(f"Result numbers: {result_numbers}")

    # Convert back to text
    decrypted_text = numbers_to_text(result_numbers)
    return decrypted_text
```

```python
def test_example_from_assignment():
    """Test the example from the assignment"""
    print("=" * 60)
    print("TESTING EXAMPLE FROM ASSIGNMENT")
    print("=" * 60)

    # Example from the assignment
    text = "ПРИКАЗ"
    gamma = "ГАММА"

    print(f"Original text: {text}")
    print(f"Gamma: {gamma}")
    print()

    # Encrypt
    encrypted = gamma_encrypt(text, gamma)
    print(f"\nEncrypted: {encrypted}")

    # Decrypt
    decrypted = gamma_decrypt(encrypted, gamma)
    print(f"\nDecrypted: {decrypted}")

    # Expected result from assignment (for reference only)
    expected = "УСХЧБЛ"
    print(f"\nExpected (from assignment): {expected}")
    print(f"Note: Different letter numbering gives different result")

    # Verify that encryption and decryption work correctly
    success = text.lower() == decrypted
    print(f"\nEncryption/Decryption successful: {success}")

    print("\n" + "=" * 60)
```

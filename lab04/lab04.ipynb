```python
def euclid_simple(a, b):
    """
    Simple Euclidean algorithm
    Input: integers a, b
    Output: d = GCD(a, b)
    """
    while a != 0 and b != 0:
        if a >= b:
            a = a % b
        else:
            b = b % a
    return a or b
```

```python
def euclid_binary(a, b):
    """
    Binary Euclidean algorithm
    Input: integers a, b
    Output: d = GCD(a, b)
    """
    g = 1

    # Remove common factors of 2
    while a % 2 == 0 and b % 2 == 0:
        a //= 2
        b //= 2
        g *= 2

    u, v = a, b

    # Main loop
    while u != 0:
        # Remove factors of 2 from u
        while u % 2 == 0:
            u //= 2

        # Remove factors of 2 from v
        while v % 2 == 0:
            v //= 2

        # Subtract smaller from larger
        if u >= v:
            u = u - v
        else:
            v = v - u

    # Multiply back the common factors of 2
    d = g * v
    return d
```

```python
def euclid_extended(a, b):
    """
    Extended Euclidean algorithm
    Input: integers a, b
    Output: (d, x, y) such that d = GCD(a, b) and a*x + b*y = d
    """
    if a == 0:
        return (b, 0, 1)

    d, x1, y1 = euclid_extended(b % a, a)
    x = y1 - (b // a) * x1
    y = x1

    return (d, x, y)
```

```python
def euclid_binary_extended(a, b):
    """
    Extended binary Euclidean algorithm
    Input: integers a, b
    Output: (d, x, y) such that d = GCD(a, b) and a*x + b*y = d
    """
    g = 1

    # Remove common factors of 2
    while a % 2 == 0 and b % 2 == 0:
        a //= 2
        b //= 2
        g *= 2

    u, v = a, b
    A, B, C, D = 1, 0, 0, 1

    # Main loop
    while u != 0:
        # Remove factors of 2 from u
        while u % 2 == 0:
            u //= 2
            if A % 2 == 0 and B % 2 == 0:
                A //= 2
                B //= 2
            else:
                A = (A + b) // 2
                B = (B - a) // 2

        # Remove factors of 2 from v
        while v % 2 == 0:
            v //= 2
            if C % 2 == 0 and D % 2 == 0:
                C //= 2
                D //= 2
            else:
                C = (C + b) // 2
                D = (D - a) // 2

        # Subtract smaller from larger
        if u >= v:
            u = u - v
            A = A - C
            B = B - D
        else:
            v = v - u
            C = C - A
            D = D - B

    # Final result
    d = g * v
    x, y = C, D

    return (d, x, y)
```

```python
def test_examples():
    """Test all algorithms with examples from the assignment"""

    print("=" * 60)
    print("LABORATORY WORK No. 4")
    print("Testing GCD Algorithms")
    print("=" * 60)

    # Test cases from the assignment
    test_cases = [
        (12345, 24690),
        (12345, 54321),
        (12345, 12541),
        (91, 105),
        (999, 99)
    ]

    for a, b in test_cases:
        print(f"\na = {a}, b = {b}")
        print("-" * 40)

        # Simple Euclidean
        d1 = euclid_simple(a, b)
        print(f"Simple Euclidean: GCD = {d1}")

        # Binary Euclidean
        d2 = euclid_binary(a, b)
        print(f"Binary Euclidean: GCD = {d2}")

        # Extended Euclidean
        d3, x3, y3 = euclid_extended(a, b)
        print(f"Extended Euclidean: GCD = {d3}, x = {x3}, y = {y3}")
        print(f"  Verification: {a}*{x3} + {b}*{y3} = {a*x3 + b*y3}")

        # Extended Binary Euclidean
        d4, x4, y4 = euclid_binary_extended(a, b)
        print(f"Extended Binary Euclidean: GCD = {d4}, x = {x4}, y = {y4}")
        print(f"  Verification: {a}*{x4} + {b}*{y4} = {a*x4 + b*y4}")

        # Verify all results match
        if d1 == d2 == d3 == d4:
            print("✓ All algorithms give the same result")
        else:
            print("✗ Results do not match!")
```

```python
def main():
    """Main program function"""
    test_examples()

    print("\n" + "=" * 60)
    print("Program finished")

if __name__ == "__main__":
    main()
```

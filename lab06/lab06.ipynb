```python
def gcd(a, b):
    """
    Euclidean algorithm for greatest common divisor
    """
    while b != 0:
        a, b = b, a % b
    return a
```

```python
def pollard_rho(n, c=1):
    """
    Pollard's Rho algorithm for integer factorization
    n: number to factor (should be composite)
    c: initial value and constant in function f(x) = x² + c
    Returns: a non-trivial factor of n
    """
    # Handle trivial cases
    if n % 2 == 0:
        return 2
    if n % 3 == 0:
        return 3

    # Define function f(x) = x² + c (mod n)
    def f(x):
        return (x * x + c) % n

    # Initialize tortoise and hare
    a = 2
    b = 2
    d = 1

    # Main loop
    while d == 1:
        # Tortoise moves one step, hare moves two steps
        a = f(a)
        b = f(f(b))

        # Compute gcd of difference and n
        d = gcd(abs(a - b), n)

        # If we found n itself, algorithm failed
        if d == n:
            return pollard_rho(n, c + 1)

    return d
```

```python
def test_example():
    """Test the example from the assignment"""
    print("=" * 60)
    print("TESTING EXAMPLE FROM ASSIGNMENT")
    print("=" * 60)

    # Example from assignment
    n = 1359331
    c = 1

    print(f"Number to factor: n = {n}")
    print(f"Initial value: c = {c}")
    print(f"Function: f(x) = x² + {c} (mod n)")
    print()

    # Find factor
    factor = pollard_rho(n, c)

    print(f"Found factor: {factor}")
    print(f"Verification: {factor} × {n // factor} = {n}")
    print(f"Success: {factor * (n // factor) == n}")

    print("\n" + "=" * 60)
```

```python
def factor_number():
    """Interactive function to factor user input"""
    print("=" * 60)
    print("FACTORIZE A NUMBER")
    print("=" * 60)

    try:
        n = int(input("Enter number to factor: "))

        if n < 2:
            print("Number must be greater than 1")
            return

        print(f"\nFactoring {n}...")
        print("-" * 40)

        # Try Pollard's Rho
        factor = pollard_rho(n)

        if 1 < factor < n:
            print(f"Found factor: {factor}")
            print(f"Verification: {factor} × {n // factor} = {n}")
        else:
            print("Could not find a factor. Number might be prime.")

    except ValueError:
        print("Error: Please enter a valid integer")

    print("\n" + "=" * 60)
```
